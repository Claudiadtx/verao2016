## Desafio: Calcule o gradiente total horizontal

```python
dx = transform.derivx(x, y, anom, shape)
dy = transform.derivy(x, y, anom, shape)
thd = np.sqrt(dx**2 + dy**2)
```

```python
plt.tricontourf(y, x, thd, 60, cmap='RdBu_r')
plt.colorbar()
```
